## Numpy Arrays


### 3D-data

```python
brain = skimage.data.brain()
print(brain.shape)
```

```python
def display_slice(nz):
    plt.imshow(brain[nz])
    plt.show()
```

```python
from ipywidgets import interact, widgets

interact(display_slice, nz=widgets.IntSlider(min=0, max=brain.shape[0]-1, step=1, value=0))
```

```python
import plotly.express as px

px.imshow(brain, animation_frame=1, binary_string=True)
```

### Konvertere mellom datatyper

Ofte kan vi holde oss til datatypen vi tilfeldigvis allerede har. Men til tider er det nøvendig å endre datatype. Da er det viktig at vi går frem på riktig måte. Merk forøvrig at dersom vi går fra en datatype som bruker mange bits til en som bruker færre, f.eks uint32 til uint8 mister vi i utgangspunktet informasjon. 

Generelt sett kan vi si at vi må være ganske forsiktige når vi konverterer mellom datatyper, ettersom at vi kan potensielt miste informasjon. Eller skape endringer til datasettet som ikke er tiltenkt. Det er derimot ofte viktig påæø

Her har `skimage` innebygde funksjoner for å konvertere mellom datatyper. 
* `img_as_ubyte` (`np.uint8`)
* `img_as_uint` (`np.uint16`)
* `img_as_float` (

NB: Merk at numpy arrays allerede har en innebygget metode som heter `.astype`. Vi kan f.eks skrive `mat.astype(np.uint8)` og få en kopi av `mat`-arrayet som er `uint8`. Derimot er dette *ikke* anbefalt å gjøre med bilder, fordi transformasjonen ikke tar hensyn til informasjonen lagret i bildet som f.eks kontrast.

# Thresholding



```python
img = skimage.data.page()
plt.imshow(img, cmap='gray')
plt.show()
```

```python
print(img.dtype)
print(img.shape)
print(img.min())
print(img.max())
```

```python
def plot_img_with_given_threshold(threshold):
    plt.imshow(img < threshold, cmap='gray')
    plt.show()
```

```python
interact(plot_img_with_given_threshold, threshold=widgets.IntSlider(min=img.min(), max=img.max()))
```
